Write a query to return the furthest-separated pair of cities for each state, and the corresponding distance (in degrees, rounded to 2 decimal places) between those two cities.

# Part A

Write a query that compares every city to every other city, with no duplicates or reversals.

### Example answer

````python
%%sql

SELECT
    s1.state, 
    s1.city AS city1, 
    s2.city AS city2,
    s1.latitude AS city1_lat, 
    s1.longitude AS city1_long, 
    s2.latitude AS city2_lat, 
    s2.longitude AS city2_long
FROM stations s1
JOIN stations s2
    ON s1.state = s2.state
    AND s1.city < s2.city
````

# Part B

Using the subquery from Part A, compute the distance between each pair of cities.


### Example answer

````python
%%sql

WITH city_pairs AS (
    SELECT
        s1.state, 
        s1.city AS city1, 
        s2.city AS city2,
        s1.latitude AS city1_lat, 
        s1.longitude AS city1_long, 
        s2.latitude AS city2_lat, 
        s2.longitude AS city2_long
    FROM stations s1
    JOIN stations s2
        ON s1.state = s2.state
        AND s1.city < s2.city
)

SELECT 
    *,
    round( ((city1_lat - city2_lat)^2 + 
            (city1_long - city2_long)^2) ^ 0.5, 2) AS dist
FROM city_pairs
````

Using the subqueries from Parts A & B, sort the results by the computed distance.

### Example answer

````python
%%sql

WITH city_pairs AS (
    SELECT 
        s1.state, 
        s1.city AS city1, 
        s2.city AS city2, 
        s1.latitude AS city1_lat, 
        s1.longitude AS city1_long, 
        s2.latitude AS city2_lat, 
        s2.longitude AS city2_long
    FROM stations s1
    JOIN stations s2
    ON s1.state = s2.state
    AND s1.city < s2.city
),

-- add a column displaying rounded Euclidean distance

distances AS ( 
    SELECT 
        *,
        round( ((city1_lat - city2_lat)^2 + (city1_long - city2_long)^2) ^ 0.5, 2) AS dist
    FROM city_pairs
)

-- rank each city pair by descending distance for each state

SELECT
    *, 
    rank() OVER (PARTITION BY state ORDER BY dist DESC) AS dist_rank
FROM distances
````

## The full solution is given below

````python
%%sql

-- self-join on matching states and city < city 
-- (avoids identical and double-counted city pairs), 
-- pulling state, city pair, and latitude/longitude 
-- coordinates for each city

WITH city_pairs AS (
    SELECT 
        s1.state, 
        s1.city AS city1, 
        s2.city AS city2,
        s1.latitude AS city1_lat, 
        s1.longitude AS city1_long, 
        s2.latitude AS city2_lat, 
        s2.longitude AS city2_long
    FROM stations s1
    JOIN stations s2
    ON s1.state = s2.state
    AND s1.city < s2.city
),

-- add a column displaying rounded Euclidean distance

distances AS ( 
    SELECT *,
            round( ((city1_lat - city2_lat)^2 + (city1_long - city2_long)^2) ^ 0.5, 2) AS dist
    FROM city_pairs
),

-- rank each city pair by descending distance for each state

dist_ranks AS (
    SELECT *, 
            rank() OVER (PARTITION BY state ORDER BY dist DESC) AS dist_rank
    FROM distances
)

-- return the city pair with maximium separation

SELECT state, city1, city2, dist 
FROM dist_ranks
WHERE dist_rank = 1
````
